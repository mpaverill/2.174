# Ignition Classification

Notebook for building and evaluating an ignition classifier.


```python
import os

import pandas as pd

# Load data (expects to be run from the Final_Project folder)
data_path = "class_data.csv"
if not os.path.exists(data_path):
    raise FileNotFoundError(
        f"Could not find {os.path.abspath(data_path)}. Current working dir: {os.getcwd()}"
    )

df = pd.read_csv(data_path)

print("shape:", df.shape)
display(df.head())
```

## Clean features 

```python
# Remove unwanted features (start simple, then iterate)
# Keep this list editable as we refine feature selection.
unwanted_features = [
    "Test File", "Material", "Coefficient of Restitution", "Output Velocity (m/s)"  # identifier; usually not a generalizable predictor
]

target_cols = ["Flash", "Ignition"]

# Drop only columns that actually exist
cols_to_drop = [c for c in unwanted_features if c in df.columns]
df_model = df.drop(columns=cols_to_drop).copy()

print("Dropped columns:", cols_to_drop)
print("Modeling dataframe shape:", df_model.shape)
print("Modeling columns:")
print(df_model.columns.tolist())

# Convenience splits for the next step
X_base = df_model.drop(columns=target_cols)
y_flash = df_model["Flash"]
y_ignition = df_model["Ignition"]

print("\nX_base shape:", X_base.shape)
print("y_flash positive rate:", y_flash.mean())
print("y_ignition positive rate:", y_ignition.mean())
```
